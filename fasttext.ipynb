```python

import re
import spacy

nlp = spacy.load('ru_core_news_sm')  

def preprocess_text(text):
    text = re.sub(r'[«"\'»,).;:]', '', text)
    text = ' '.join(text.split())
    
    doc = nlp(text)
    lemmatized_text = ' '.join([token.lemma_ for token in doc])
    
    return lemmatized_text

input_file = '/kaggle/input/traintext/train.txt'

with open(input_file, 'r', encoding='utf-8') as infile, open('cleaned_train.txt', 'w', encoding='utf-8') as outfile:
    for line in infile:
        cleaned_line = preprocess_text(line)
        outfile.write(cleaned_line + '\n')

print("Обработка завершена. Очищенные и лемматизированные данные сохранены в", output_file)

```
